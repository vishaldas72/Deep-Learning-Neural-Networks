```python
import numpy as np
```

```python
yhat = np.array([0,1,1,1,0,1,0])
y  = np.array([0.1,0.6,1,0.9,0.2,0.8,0])
```

```python
def mae(yhat,y):
    total_error = 0
    for yp,yt in zip(yhat,y):
        total_error += abs(yt-yp)
    print('total_error:',total_error)
    mae = total_error/len(y)
    print('mae:',mae)
    return mae
```

```python
mae(yhat,y)
```

The above expressions can be put together into a function

```python
def log_loss(y,yhat):
    epsilon = 1e-10
    yhat_new = [max(i,epsilon) for i in yhat]
    yhat_new = [min(i,1-epsilon) for i in yhat_new]
    yhat_new = np.array(yhat_new)
    return (-np.mean(y*np.log(yhat_new)+(1-y)*np.log(1-yhat_new)))
```

```python
log_loss(y,yhat)
```

```python
y_predicted = np.array([1,1,0,0,1])
y_true = np.array([0.30,0.7,1,0,0.5])
```

```python
def mse(y_true,y_predicted):
    total_error = 0
    for yt,yp in zip(y_true,y_predicted):
        total_error += (yt-yp)**2
    print('Total Squared Error is :',total_error)
    mae = total_error/len(y_true)
    print('MAE',mae)
    return mse
```

```python
mse(y_predicted,y_true)
```
